## _Introduction_

Newspapers, magazines, and other forms of print media used to be the primary way that people would consume news and media. However, due to advances within the technological age, electronic media has taken over the news flow. People now get alerts on their phones, breaking news updates on their laptops, and notifications on their tablets. Articles are scrolled through and opinion pieces are debated in the comments section.

However, this shift from paper to electronic does not indicate the death of print media. Newspapers like the _New York Times_ and the _Washington Post_ both still circulate hundreds of thousands of issues a day. Magazines like _People_ and _Sports Illustrated_ have each circulated millions of copies. Print media is still being produced, and people are still consuming it.

In fact, print media is doing so well that [Pulitzer Prize](https://www.pulitzer.org/) worthy pieces are being written every day. Pulitzer Prizes showcase excellence in print journalism, recognizing masterfully written pieces in the media industry. The fact that these pieces are still prominent displays how relevant print media still is today.

So, how will the rise of electronic media affect the long established industry of print media? Will newspapers, magazines, and the like slowly decline into oblivion? Or will the work of Pullitzer Prize winners and other columnists carry print media throughout this electronic age?

In this project, I will attempt to tackle some of these questions.

## Research Questions

To try and properly understand the future of print media, I will look at the following research questions.

### How has the number of Pulitzer Prize winners affected newspaper circulation?

One of the prime accomplishments in modern print journalism is receiving a Pulitzer Prize. These awards indicate quality journalism, as receipients are often held to a higher standard of writing. If a correlation can be established between the number of Pulitzer Prize winners a newspaper has and the circulation of said newspaper, companies could focus on quality of pieces as opposed to quantity of work. Print media companies could focus on creating more detailed, Pulitzer-quality work that would draw readers to their newspaper, consequently increasing profits.

### How have revenues from advertising and circulation changed over time?

The two primary ways for print newspapers to obtain revenue is by advertising and circulation. If it can be observed that revenue via one of these methods increases or decreases over time, then newspaper companies can make the proper adjustments to focus their revenue efforts on one of these streams. This would help with the proper allocation of resources, such as employees, time, and financial efforts.

### How has the number of Pulitzer Prize winners affected newspaper circulation?

For this analysis, I am seeking to determine if there is a correlation between the number of Pulitzer Prize winners and finalists a newspaper has and how high or low its daily circulation is. Therefore, I will primarily utilize FiveThirtyEight's dataset.

The first step I can do is to create a correlation matrix. It is important to note that this will not indicate causation. Rather, it could be the first step in establishing a more concrete correlation.

```python
# Creating a correlation matrix
fte_dataset.corr()
```

The correlation between *Pulitzer Prize Winners and Finalists, 1990-2003* and *Daily Circulation, 2004* is 0.446074. This is not necessarily a strong correlation, but it is definitely not a weak one either. It tells me that there is something here worth investigating.

The correlation between *Pulitzer Prize Winners and Finalists, 2004-2014* and *Daily Circulation, 2013* is 0.532106. Again, this is not necessarily a strong correlation, but it is definitely not a weak one either. It tells me that there is something worth investigating here as well.

The correlation between *Total Pulitzer Prize Winners and Finalists, 1990-2014* and *Total Circulation, 2004 and 2013* is 0.497523. Once again, this is not necessarily a strong correlation, but it is definitely not a weak one either. It tells me that there is something here worth investigating too.

In order to properly determine a correlation, I must run a regression. I have chosen to run a linear regression. Below is a plot of a regression line on a plot comparing *Pulitzer Prize Winners and Finalists, 1990-2003* to *Daily Circulation, 2004*.

```python
# Defining a figure
plt.figure()

# Plotting the regression line
sns.regplot(fte_dataset['Pulitzer Prize Winners and Finalists, 1990-2003'], fte_dataset['Daily Circulation, 2004'])

# Adding a title to the plot
plt.title('Pulitzer Prizes vs. Daily Circulation')
```

Due to the scale in the plot above, most of the values fall towards the smaller values on the plot. Again, I should note the use of scientific notation on the y axis.

Now, in order to properly perform the linear regression, I will need to import some functionalities from the statsmodels library.

```python
# Importing statsmodels and associated functions
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

Now that I have imported the necessary libraries and functions, I can utilize them to perform a linear regression.

```python
# Running the regression
pul_circ_reg_results = smf.ols('Q("Daily Circulation, 2004") ~ Q("Pulitzer Prize Winners and Finalists, 1990-2003")', data = fte_dataset).fit()
print(pul_circ_reg_results.summary())
```

As can be seen from the output above, the R-squared for this model is only 0.199, or 19.9%. This is not a very high R-squared at all, which tells me this variable alone can not properly determine newspaper circulation. Consequently, according to this model, I can not use Pulitzer Prizes to determine newspaper circulation.

But what if I use *Daily Circulation, 2013* and *Pulitzer Prize Winners and Finalists, 2004-2014*? Will that model produce better results?

First, I will plot the regression line.

```python
# Defining a figure
plt.figure()

# Plotting the regression line
sns.regplot(fte_dataset['Pulitzer Prize Winners and Finalists, 2004-2014'], fte_dataset['Daily Circulation, 2013'])

# Adding a title to the plot
plt.title('Pulitzer Prizes vs. Daily Circulation')
```

```python
# Running the regression
pul_circ_reg_results_two = smf.ols('Q("Daily Circulation, 2013") ~ Q("Pulitzer Prize Winners and Finalists, 2004-2014")', data = fte_dataset).fit()
print(pul_circ_reg_results_two.summary())
```

Here, I see that the R-squared is 0.283, or 28.3%. While this R-squared is definitely better than the previous model's R-squared, it is still not nearly high enough to say that this variable alone can properly determine newspaper circulation. Consequently, according to this model, I can not use Pulitzer Prizes to determine newspaper circulation.

Finally, what if I use *Total Circulation, 2004 and 2013* and *Total Pulitzer Prize Winners and Finalists, 1990-2014*? Will that model produce better results?
